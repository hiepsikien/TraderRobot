```python
import talib as ta
import pandas as pd
from my_logging import logger
filepath = "bitcoin.csv"
```

```python
raw = pd.read_csv(filepath, parse_dates=["Date"],index_col="Date") 

```

```python
open = raw[["Open"]].to_numpy().flatten()
close = raw[["Close"]].to_numpy().flatten()
high = raw[["High"]].to_numpy().flatten()
low = raw[["Low"]].to_numpy().flatten()
volume = raw[["Volume"]].to_numpy().flatten()
```

```python
obv = ta.OBV(raw.Close,raw.Volume)
raw["obv"] = obv
raw
```

```python
mfi = ta.MFI(high = high, low = low, close=close, volume=volume, timeperiod=14)
```
